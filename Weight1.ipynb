```r
library(KFAS)
```

# Data

```r
# read data
Weight <- ts(scan("leadingDAT_src/Weight.dat"))
```

# Model Definition&Estimation

```r
# local model definition
mod <- SSModel(Weight ~ SSMtrend(1,Q=NA),H=NA)
```

### Note:
* Q: variance of state
* H: variance of observation

```r
# estimation
fit <- fitSSM(mod,numeric(2),method="BFGS")
```

BFGS is fast, but not stable??  
if doesn't work, another method should be tried.

```r
# perform kalman filtering & smoothing
kfs <- KFS(fit$model)
```

# Kalman Filtering

```r
# filtered information
afilt <- kfs$att
Pfilt <- kfs$Ptt[,,]
afiltupper <- afilt + sqrt(Pfilt) * qnorm(.975)
afiltlower <- afilt - sqrt(Pfilt) * qnorm(.975)
```

```r
plot(Weight,type="o",lty=3,xlab="Elapsed Days",ylab="Weight[kg]",ylim=c(83,87))
lines(afilt,lwd=2)
lines(afiltupper)
lines(afiltlower)
```

# Kalman Smoothing

```r
# smoothed information
alphahatconf <- predict(fit$model,interval="confidence",level=.95)
```

```r
plot(Weight,type="o",lty=3,xlab="Elapsed Days",ylab="Weight[kg]",ylim=c(83,87))
lines(alphahatconf[,1],lwd=2)
lines(alphahatconf[,2])
lines(alphahatconf[,3])
```

# NA Compensation

```r
# fill 21~40th days as NA
WeightNA <- Weight[c(1:20,rep(NA,20),41:60)]
modNA <- SSModel(WeightNA ~ SSMtrend(1,Q=NA),H=NA)
fitNA <- fitSSM(modNA,numeric(2),method="BFGS")
confNA <- predict(fitNA$model,interval="confidence",level=.95)
preNA <- predict(fitNA$model,interval="prediction",level=.95)
```

```r
plot(Weight,type="o",lty=3,xlab="Elapsed Days",ylab="Weight[kg]",ylim=c(83,87))
lines(21:40,Weight[21:40],type="o",lty=3,col=8)
lines(confNA[,1],lwd=2)
lines(confNA[,2])
lines(confNA[,3])
lines(21:40,preNA[21:40,2],lty=2)
lines(21:40,preNA[21:40,3],lty=2)
```

### Note:
* solid line: state confidence interval
* dashed line: observation prediction interval

# Long-term Forecast

```r
mod50 <- SSModel(Weight[1:50] ~ SSMtrend(1,Q=NA),H=NA)
fit50 <- fitSSM(mod50,numeric(2),method="BFGS")
conf50 <- predict(fit50$model,interval="confidence",n.ahead=10)
pre50 <- predict(fit50$model,interval="prediction",n.ahead=10)
```

```r
plot(Weight,type="o",lty=3,xlab="Elapsed Days",ylab="Weight[kg]",ylim=c(83,87))
lines(51:60,Weight[51:60],type="o",lty=3,col=8)
lines(51:60,conf50[,1],lwd=2)
lines(51:60,conf50[,2])
lines(51:60,conf50[,3])
lines(51:60,pre50[,2],lty=2)
lines(51:60,pre50[,3],lty=2)
```
