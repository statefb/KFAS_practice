```r
library(KFAS)
source("./practice//kfas_helper.r")
```

```r
sales <- read.csv("leadingDAT_src/sales.csv",fileEncoding="shift-jis")
```

```r
head(sales)
```

```r
fabric <- ts(sales[,2])
# NA = to be predicted (2 years)
fabricNA <- c(fabric[1:120],rep(NA,24))
```

```r
# Model Definition
mod1 <- SSModel(fabricNA ~ SSMtrend(1,Q=NA) + SSMseasonal(12,Q=0),H=NA)
mod2 <- SSModel(fabricNA ~ SSMtrend(1,Q=NA) + SSMseasonal(12,Q=NA),H=NA)
mod3 <- SSModel(fabricNA ~ SSMtrend(2,Q=list(0,NA)) + SSMseasonal(12,Q=0),H=NA)
mod4 <- SSModel(fabricNA ~ SSMtrend(2,Q=list(0,NA)) + SSMseasonal(12,Q=NA),H=NA)
```

```r
# Params Estimation
fit1 <- fitSSM(mod1,numeric(2),method="BFGS")
fit2 <- fitSSM(mod2,numeric(3),method="BFGS")
fit3 <- fitSSM(mod3,numeric(2),method="BFGS")
fit4 <- fitSSM(mod4,numeric(3),method="BFGS")
```

```r
# Kalman filtering
kfs1 <- KFS(fit1$model)
kfs2 <- KFS(fit2$model)
kfs3 <- KFS(fit3$model)
kfs4 <- KFS(fit4$model)
```

```r
# Likelihood
logLik1 <- kfs1$logLik - sum(kfs1$Finf>0) * log(2*pi)/2
logLik2 <- kfs2$logLik - sum(kfs2$Finf>0) * log(2*pi)/2
logLik3 <- kfs3$logLik - sum(kfs3$Finf>0) * log(2*pi)/2
logLik4 <- kfs4$logLik - sum(kfs4$Finf>0) * log(2*pi)/2
```

```r
# AIC 
AIC1 <- -2*logLik1 + 2*( 2 + 12 )
AIC2 <- -2*logLik2 + 2*( 3 + 12 )
AIC3 <- -2*logLik3 + 2*( 2 + 13 )
AIC4 <- -2*logLik4 + 2*( 3 + 13 )
```

```r
print(AIC1);print(AIC2);print(AIC3);print(AIC4);
```

```r
# confirm whether wrapper generates the same results

mods <- list(
    SSModel(fabricNA ~ SSMtrend(1,Q=NA) + SSMseasonal(12,Q=0),H=NA),
    SSModel(fabricNA ~ SSMtrend(1,Q=NA) + SSMseasonal(12,Q=NA),H=NA),
    SSModel(fabricNA ~ SSMtrend(2,Q=list(0,NA)) + SSMseasonal(12,Q=0),H=NA),
    SSModel(fabricNA ~ SSMtrend(2,Q=list(0,NA)) + SSMseasonal(12,Q=NA),H=NA)
)

# param estimation
fits <- lapply(mods,fit_wrapper,method="BFGS")
mods.fit <- get_models(fits)
# kalman filter
kfs <- lapply(mods.fit,KFS,smoothing=c("state","mean","disturbance"))
# prediction(confidence)
confs <- lapply(mods.fit,predict,interval="confidence",level=.95)
# error analysis
errors <- lapply(kfs,get_standardized_error)

# Log likelihood
lls <- lapply(kfs,calc_log_likelihood)
# AIC
aics <- calc_AICs(kfs,mods)
# select the best
best_idx <- select_best_model(kfs,mods)$best_index
best_mod <- select_best_model(kfs,mods)$model
# 
print(paste0("log likelihood: ",lls))
print(paste0("AIC: ",aics))
print(paste0("best model index: ",best_idx))
```

2nd model (local level & dynamic seasonal model) is the best

# Calender Effect
include calendar effect & leap year effect

```r
dates <- seq(as.Date("2002-01-01"),as.Date("2013-12-31"),by=1)
weeks <- table(substr(dates,1,7),weekdays(dates,T))
sun <- weeks[,"日"]
mon <- weeks[,"月"]-sun; tue <- weeks[,"火"]-sun; wed <- weeks[,"水"]-sun
thu <- weeks[,"木"]-sun; fry <- weeks[,"金"]-sun; sat <- weeks[,"土"]-sun
calendar <- cbind(mon, tue, wed, thu, fry, sat)

leapyear  <- rownames(weeks) %in% c("2004-02","2008-02","2012-02")
```

```r
head(weeks)
print(grep("2004-02",rownames(weeks)))
head(calendar)
print(leapyear)
```

```r
# model
modCalendar <- SSModel(fabric ~ SSMtrend(2,Q=list(0,NA))
                      + SSMseasonal(12,sea.type="dummy")
                      + leapyear + calendar, H=NA
                      )
fitCalendar <- fitSSM(modCalendar,numeric(2),method="BFGS")
kfsCalendar <- KFS(fitCalendar$model)
```

```r
# goodness of the model
logLikCalendar <- kfsCalendar$logLik-sum(kfsCalendar$Finf>0)*log(2*pi)/2
AICCalendar <- -2*logLikCalendar + 2*( 2 + 20 )
print(AIC1);print(AIC2);print(AIC3);print(AIC4);
print(AICCalendar)
print(logLik1);print(logLik2);print(logLik3);print(logLik4);print(logLikCalendar)
```

```r
plot(kfsCalendar$muhat - kfsCalendar$alphahat[,"level"], type="l", col=4,
  xaxs="i", xaxt="n", xlab="", ylab="sales")
axis(side=1,at=1+0:11*12,labels=c("02/1","03/1","04/1","05/1","06/1","07/1","08/1","09/1","10/1","11/1","12/1","13/1"))
abline(v=120.5,lty=3)
```
