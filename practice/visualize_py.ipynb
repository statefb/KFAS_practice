## データ読み込み

```python
df = pd.read_csv("number_of_guests.csv",encoding="shift-jis")
print("data frame size: {}".format(df.shape))
df.index = jpd.to_datetime(df["年月"])  # 年月をインデックスへ
df = df.iloc[:,1:]  # 年月列を削除
df.head()
```

```python
# 平均訪問者数
df.mean().iloc[1:].sort_values(ascending=False).iplot()
```
