```r
library(KFAS)
```

```r
Weight <- ts(scan("leadingDAT_src/Weight.dat"))
```

# Comparison of Models
1. local level model  
level walks randomly
2. smoothed trend model  
slope of level walks randomly
3. local linear trend model  
both of them walk randomly

```r
mod1 <- SSModel(Weight ~ SSMtrend(1,Q=NA),H=NA)
mod2 <- SSModel(Weight ~ SSMtrend(2,Q=list(0,NA)),H=NA)
mod3 <- SSModel(Weight ~ SSMtrend(2,Q=list(NA,NA)),H=NA)
```

params to be estimated are defined as "NA".

```r
fit1 <- fitSSM(mod1,numeric(2),method="BFGS")
fit2 <- fitSSM(mod2,numeric(2),method="BFGS")
fit3 <- fitSSM(mod3,numeric(3),method="BFGS")
```

```r
kfs1 <- KFS(fit1$model)
kfs2 <- KFS(fit2$model)
kfs3 <- KFS(fit3$model)
```

```r
# Maximam log likelihood
logLik1 <- kfs1$logLik - sum(kfs1$Finf>0) * log(2*pi)/2
logLik2 <- kfs2$logLik - sum(kfs2$Finf>0) * log(2*pi)/2
logLik3 <- kfs3$logLik - sum(kfs3$Finf>0) * log(2*pi)/2
```

second term: modify term (Durbin and Koopman,2012)

```r
print(logLik1);print(logLik2);print(logLik3);
```

```r
# AIC
AIC1 <- -2*logLik1 + 2*(2+1)
AIC2 <- -2*logLik2 + 2*(2+2)
AIC3 <- -2*logLik3 + 2*(3+2)
```

AIC = -2 \* maximum likelihood +  
2 \* (number of unknown params + number of dynamic states)

```r
print(AIC1); print(AIC2); print(AIC3)
```

local level model is the best model.

```r
par(mfrow=c(2,1))
par(mar=c(2,4,1,1))
plot(Weight,type="o",lty=3)
lines(kfs1$alphahat[,"level"],lwd=2,col=8)
lines(kfs2$alphahat[,"level"],lwd=2)
lines(kfs3$alphahat[,"level"],lwd=2,lty=5)

plot(kfs2$alphahat[,"slope"],type="l",lwd=2,xlab="",ylab="slope")
lines(kfs3$alphahat[,"slope"],lwd=2,lty=5)
```
