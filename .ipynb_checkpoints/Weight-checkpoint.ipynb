```r
library(KFAS)
```

# Reading Data

```r
# read weight data
Weight <- ts(scan("data/Weight.dat"))
```

```r
plot(Weight)
```

# Local Model Definition

```r
# definition of local model
# Note: substitute NA into unknown parameters in advance
mod <- SSModel(Weight ~ SSMtrend(1,Q=NA),H=NA)
# maximum likelihood-estimation for Q,H (unknown parameters)
# Note: numeric(2): c(0,0) initial value
fit <- fitSSM(mod,numeric(2),method="BFGS")

print(fit$model)
```

# Kalman Filtering and Smoothing

```r
# perform kalman filtering and smoothing
kfs <- KFS(fit$model)
```

```r
# extract filtered data
afilt <- kfs$a[-1]
Pfilt <- kfs$P[,,-1] - fit$model$Q
afiltconf <- cbind(afilt+sqrt(Pfilt)*qnorm(0.025),afilt+sqrt(Pfilt)*qnorm(0.975))
```

```r
plot(Weight,type="o",lty=3,xlab="elapsed days",ylab="weight[kg]",ylim=c(83,87))
lines(afilt,lwd=2)
lines(afiltconf[,1])
lines(afiltconf[,2])
```

```r
# extract smoothed data
alphahatconf <- predict(fit$model,interval="confidence",level=.95)
```

```r
plot(Weight,type="o",lty=3,xlab="elapsed days",ylab="weight[kg]",ylim=c(83,87))
lines(alphahatconf[,1],lwd=2)
lines(alphahatconf[,2])
lines(alphahatconf[,3])
```
